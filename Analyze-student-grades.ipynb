```python
#In this code, we want to analyze students' grades based on the instructions given.
#First, we should read an Excel file that contains students' grades.
#Then, calculating the average.
#Next, determine the letter grades based on the final grade.
#Then, adding the letter grade as a column.
# Finally, saving the new sheet.

import pandas as pd

# Reading the Excel file
df = pd.read_excel("Grades_Short.xlsx")

# Calculating the average
df["Final_Grade"] = df[["Assignment_1", "Assignment_2", "Quiz_1", "Quiz_2", "Mid_Term_Exam", "Final_Exam"]].mean(axis=1)

# Letter grade based on the final grades.
def get_letter_grade(score):
    if score > 90:
        return "A+"
    elif 90>= score > 80:
        return "A"
    elif 80>= score > 70:
        return "B"
    elif 70>= score > 60:
        return "C"
    elif 60>= score > 55:
        return "D"
    else:
        return "F"

# Adding a new column for letter grades
df["Letter_Grade"] = df["Final_Grade"].apply(get_letter_grade)

# New CSV file
df.to_csv("Grades_Modified.csv", index=False)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df
```
